```python
dt[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]=dt[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
dt.info()
```

```python
dt.to_csv('mix_cleaned.csv',index=False)
```

```python
dtc=dt.drop(['day','month','year'],axis=1)
```

```python
dtc.head()
```

```python
dtc['Classes'].value_counts()
```

```python
dtc['Classes']=dtc['Classes'].str.strip()
dtc['Classes'].value_counts()
```

```python
dtc['Classes']=np.where(dtc['Classes'].str.contains('not fire'),0,1)
```

```python
dtc.head()
```

```python
dtc.tail()
```

```python
dtc['Classes'].value_counts()
```

```python
percentage=dtc['Classes'].value_counts(normalize=True)*100 #percentage for pie chart
```

```python
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```
